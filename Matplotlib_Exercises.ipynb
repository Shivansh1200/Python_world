```python
#3. Draw a line using given axis values taken from a text file, with suitable label in the X axis and the Y-axis and a title.
'''test.txt
12
24
31'''

#solution:
import matplotlib.pyplot as plt
# opening a given file
with open('test.txt') as f:
    data=f.read()
    data=data.split('\n')
    x= [row.split('')[0] for row in data]
    y=[row.split('')[1] for row in data]
    plt.plot(x, y)
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.title('Sample graph!')
    plt.show()
```

```python
# python code for generating a histogram
from matplotlib import pyplot as plt
population_ages=[22,33,44,55,66,77,88,99,23,25,27,29,34,36,38,80,90,79,100]
plt.hist(population_ages, bins=5, histtype='bar')
plt.xlabel('Years')
plt.ylabel('Population growth')
plt.title('population age analysis graph')
plt.legend()
plt.show()
```

```python
# adding style to graph
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
X=[5,8,10]
Y=[12,16,6]

X2=[6,9,11]
Y2=[6,15,7]
plt.plot(X,Y,'g',label='line-One', linewidth=5)
plt.plot(X2, Y2,'c',label='line_two', linewidth=5)
plt.title('Epic info.')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True, color='k')
plt.show()
```
